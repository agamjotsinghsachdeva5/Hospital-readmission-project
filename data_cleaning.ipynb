```python
import pandas as pd
df=pd.read_csv("final_data1.csv")
```

```python
agg_dict = {col: "first" for col in df.columns if col not in ["diag_category", "icd10_code", "diag_desc"]}
agg_dict["diag_category"] = lambda x: list(set(x.dropna()))

df_adm = df.groupby("admission_id", as_index=False).agg(agg_dict)

# Now df_adm["diag_category"] is a list of categories per admission
print(df_adm[["admission_id", "diag_category"]].head())
```

```python
print(len(df_adm))
```

```python
print(df_adm.columns)
```

```python
df_adm=df_adm.drop(["Unnamed: 0","patient_id","admission_id","admit_date","discharge_date",'hospital_id','name','bill_id',
       'total_cost_inr', 'govt_subsidy_inr','diag_id'],axis=1)
```

```python
print(df_adm["insurance_type"].isnull().sum())
```

```python
df_adm["insurance_type"] = df_adm["insurance_type"].fillna("None")
```

```python
known_types = ["ESI", "Private", "Ayushman", "None"]
df_adm["insurance_type"] = df_adm["insurance_type"].apply(lambda x: x if x in known_types else "Other")
```

```python
print(df_adm["insurance_type"].value_counts())
```
