```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data/patients.csv")
print(df.head())
```

```python
print(df.columns)
```

```python
print(df["gender"].value_counts().sort_values())
```

```python


# Value counts
counts = df["insurance_type"].value_counts()

# Pie chart with labels and percentages
plt.pie(
    counts,
    labels=counts.index,        # labels from the index (Male, Female, Others)
    autopct="%1.1f%%",          # format for percentages
    startangle=90,              # optional: rotate for better look
    colors=[ "lightpink", "lightgreen","skyblue","red"]  # optional: custom colors
)

plt.title("bpl_card Distribution")
plt.show()
```

```python
sns.countplot(x="state",data=df)
plt.xticks(rotation=90)
plt.xlabel("PREVIOUS ADMISSIONS ")
plt.ylabel("PATIENT COUNT")
plt.show()
```

```python


# Value counts
counts = df["prev_admissions"].value_counts()


# Pie chart with labels and percentages
plt.pie(
    counts,
    labels=counts.index,        # labels from the index (Male, Female, Others)
    autopct="%1.1f%%",          # format for percentages
    startangle=90,              # optional: rotate for better look
    colors=[ "lightpink", "lightgreen","skyblue","red"]  # optional: custom colors
)

plt.title("bpl_card Distribution")
plt.show()
```

```python
print(df["state"].value_counts().index)
```

```python
region_map = {
    "Maharashtra": "West",
    "Gujarat": "West",
    "Rajasthan": "West",
    "Punjab": "North",
    "Haryana": "North",
    "Uttar Pradesh": "North",
    "Madhya Pradesh": "Central",
    "West Bengal": "East",
    "Odisha": "East",
    "Bihar": "East",
    "Tamil Nadu": "South",
    "Kerala": "South",
    "Karnataka": "South",
    "Andhra Pradesh": "South",
    "Telangana": "South"
}

df["region"] = df["state"].map(region_map)
```

```python
region_counts = df["region"].value_counts()
# Pie chart of normalized region counts
plt.pie(
    region_counts,
    labels=region_counts.index,   # region names
    autopct="%1.1f%%",                # show percentages
    startangle=90,                    # rotate for better look
    colors=["skyblue", "lightgreen", "lightcoral", "gold", "violet"]  # optional colors
)
plt.title("Normalized Region Distribution (per state)")
plt.show()
region_percentages = region_counts * 100 / region_counts.sum()
# Print with region names
print(region_percentages)

```
